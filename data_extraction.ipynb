# Financial Analysis Summary

This analysis examines key financial metrics from the 10-K filings of Microsoft, Tesla, and Apple for the fiscal years 2021, 2022, and 2023. We calculated year-over-year percentage changes for Total Revenue, Net Income, Total Assets, Total Liabilities, and Cash Flow from Operating Activities to identify trends and shifts in company performance.


```python
df = pd.read_csv('/content/10K_data.csv')
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)': 'mean',
    'Liabilities Growth (%)': 'mean',
    'Cash Flow Growth (%)': 'mean'
}).reset_index()
print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```

## Microsoft: Consistent Performance

### Revenue Growth
Microsoft experienced average revenue growth of **7.52%**. This indicates steady and healthy growth, driven by strong product demand and the company’s continued success in expanding its cloud computing and enterprise services.

### Net Income Growth
Microsoft’s net income growth averaged **7.09%**, reflecting solid profitability improvements over the period. This shows that Microsoft has been able to translate revenue growth into profit efficiently, likely through its high-margin cloud and software services.

### Asset and Liability Growth
Microsoft’s assets grew at a healthy rate of **12.41%**, while liabilities grew at a similar pace of **7.40%**. This balance between asset and liability growth suggests that Microsoft has been expanding its operations and investing in new ventures, while maintaining a controlled level of debt.

### Cash Flow Growth
The company saw average cash flow growth of **11.24%**, indicating that its core operations are generating strong cash flows, likely driven by increased efficiency and high demand for its services. This is a positive indicator for Microsoft’s financial health.


## Apple: Modest Growth

### Revenue Growth
Apple's revenue growth over the past three years averaged **1.66%**, indicating relatively stable but modest expansion. This slow growth suggests that Apple may be facing market saturation or slowing demand for its products, making it more difficult to achieve the significant revenue growth it experienced in prior years.

### Net Income Growth
Apple’s net income growth averaged **0.87%**, reflecting very limited improvement in profitability. This aligns with the slight revenue growth, suggesting that cost control and profitability measures haven’t significantly boosted the company’s bottom line.

### Asset and Liability Growth
The growth in assets for Apple was minimal at **0.15%**, while liabilities grew slightly more at **0.36%**. These low growth rates indicate that Apple has maintained a relatively stable balance sheet, with little change in its overall assets or liabilities.

### Cash Flow Growth
Apple’s cash flow from operating activities showed an average growth of **2.64%**, which is modest but indicates that the company has managed to slightly improve its cash generation from core operations despite slow revenue and net income growth.


## Tesla: Rapid Expansion

### Revenue Growth
Tesla displayed the most aggressive revenue growth, averaging **23.38%** year-over-year. This high growth rate is indicative of Tesla’s rapid market expansion and growing demand for electric vehicles, as well as its ability to scale production.

### Net Income Growth
Tesla’s net income grew at a remarkable average rate of **47.33%**, reflecting the company’s ability to significantly improve profitability. Tesla has clearly been able to leverage its revenue growth to achieve outsized profit increases, likely due to improved production efficiencies and cost management.

### Asset and Liability Growth
Tesla’s total assets grew at a strong rate of **20.67%**, while liabilities grew at an even higher rate of **12.44%**. This shows that Tesla has been heavily investing in expanding its production capacity and operations, although it has also been taking on more liabilities to finance this growth.

### Cash Flow Growth
Tesla’s cash flow from operating activities grew by an average of **6.03%**, indicating that while the company is generating more cash, its rapid growth may be consuming a significant portion of its cash flows. Tesla’s relatively lower cash flow growth, compared to its revenue and income growth, may suggest the need for continued investment in operations.


## Conclusion

- **Apple** has demonstrated modest growth in both revenue and profitability, indicating a more mature phase in its business lifecycle with limited room for explosive growth. The slight increase in cash flow suggests that Apple’s core operations remain strong, though the company may need new growth strategies to re-accelerate.
  
- **Microsoft** has shown consistent, steady growth across all metrics. Its balanced asset and liability expansion, coupled with strong cash flow generation, suggests that Microsoft is in a stable and healthy financial position, with solid operational efficiency driving both revenue and profitability growth.

- **Tesla** is in a rapid expansion phase, with very high revenue and net income growth rates. However, Tesla’s liability growth is also substantial, reflecting its aggressive investments in capacity expansion. While cash flow is growing, the company may face challenges in sustaining this level of growth without further financial discipline.

Overall, each company demonstrates unique growth characteristics: Apple as a mature, stable company; Microsoft with steady and efficient expansion; and Tesla with rapid growth but potentially higher risk due to its reliance on external financing.
